# Исследование объявлений о продаже квартир

На анализ были предоставлены данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктов за несколько лет. Приводим к единообразию название населенных пунктов:
1. Приводим к нижнему регистру
2. Удаляем возможные пропуски в начале и конце

4. Проводим проверку


```python
df['locality_name'] = df['locality_name'].str.lower()
df['locality_name'] = df['locality_name'].str.replace('\\bпосёлок городского типа \\b', '', regex = True)
df['locality_name'] = df['locality_name'].str.replace('\\bпоселок городского типа \\b', '', regex = True)
df['locality_name'] = df['locality_name'].str.replace('\\bгородской поселок \\b', '', regex = True)
df['locality_name'] = df['locality_name'].str.replace('\\bгородской посёлок \\b', '', regex = True)
df['locality_name'] = df['locality_name'].str.replace('\\bпоселок \\b', '', regex = True)
df['locality_name'] = df['locality_name'].str.replace('\\bпосёлок \\b', '', regex = True)
df['locality_name'] = df['locality_name'].str.replace('\\bкоттеджный поселок \\b', '', regex = True)
df['locality_name'] = df['locality_name'].str.replace('\\bкоттеджный посёлок \\b', '', regex = True)
df['locality_name'] = df['locality_name'].str.replace('\\bкоттеджный \\b', '', regex = True)
df['locality_name'] = df['locality_name'].str.replace('\\bсело \\b', '', regex = True)
df['locality_name'] = df['locality_name'].str.replace('\\bдеревня \\b', '', regex = True)
df['locality_name'] = df['locality_name'].str.replace('\\bсадовое товарищество \\b', '', regex = True)
df['locality_name'] = df['locality_name'].str.replace('\\bсадоводческое некоммерческое товарищество \\b', '', regex = True)
df['locality_name'] = df['locality_name'].str.strip()
df['locality_name'].unique()
```

**Обработка пропусков kitchen_area**

Пропусков kitchen_area 2307. Возможно при заполнении карточник объекта такой информации просто не было в доступе. 
Сначало находим отношение площади кухни к общей площади, далее через цикл там где есть пропуски площадь кухни находим 
через произведение совокупной площади на долю кухни в зависимости от кол-ва комнат

```python
df['kitchen_area_ratio'] = df['kitchen_area'] / df['total_area']
df['kitchen_area_ratio'] = df['kitchen_area_ratio'].round(3)
rooms_kitchen_area_ratio = df.query('kitchen_area > 0').pivot_table(index='rooms', values='kitchen_area_ratio')
print(rooms_kitchen_area_ratio)

for room in rooms_kitchen_area_ratio.index:
    df.loc[(df['rooms'] == room) & (df['kitchen_area'].isna()), 'kitchen_area'] = df.loc[(df['rooms'] == room) & (df['kitchen_area'
                                   ].isna()), 'total_area'] * rooms_kitchen_area_ratio.loc[room, 'kitchen_area_ratio']
```

```python
df['kitchen_area'] = df['kitchen_area'].round(1)
```

```python
df['kitchen_area'] = df['kitchen_area'].fillna(0)
```

## Добавляем в таблицу новые столбцы

посчитано и добавлено в таблицу: цена одного квадратного метра

посчитано и добавлено в таблицу: день публикации объявления (0 - понедельник, 1 - вторник и т.д.)

посчитано и добавлено в таблицу: месяц публикации объявления

посчитано и добавлено в таблицу: год публикации объявления

посчитано и добавлено в таблицу: расстояние в км до центра города

```python
df['sqMeter_price'] = df['last_price'] / df['total_area'] 
df['sqMeter_price'] = df['sqMeter_price'].round(1)
df['weekday_exposition'] = df['first_day_exposition'].dt.weekday
df['month_exposition'] = df['first_day_exposition'].dt.month
df['year_exposition'] = df['first_day_exposition'].dt.year
df['cityCenters_km'] = df['cityCenters_nearest'] / 1000 
df['cityCenters_km'] = df['cityCenters_km'].round(1)
```

посчитано и добавлено в таблицу: тип этажа квартиры (значения — «первый», «последний», «другой»)

```python
def floor_category (row):
    floor = row['floor']
    floors_total = row['floors_total']
    
    try:
        if row['floor'] == 1 or row['floors_total'] == 1:
            return 'первый'
        elif row['floor'] > 1 and row['floor'] == row['floors_total']:
            return 'последний'
        else:
            return 'другой'
        
    except:
        pass
    
df['category_floor'] = df.apply(floor_category, axis=1)
```

Т.к. для строковых значений посчитать коэф корреляции не получилось ввожу колонку category_floor_numeric с  цисловыми обозначения категорий этажности:

```python
def floor_category_numeric (row):
    category = row['category_floor']
        
    try:
        if row['category_floor'] == 'первый':
            return 1
        elif row['category_floor'] == 'последний':
            return 3
        else:
            return 2
        
    except:
        pass
    
df['category_floor_numeric'] = df.apply(floor_category_numeric, axis=1)
```

```python
(df
 .pivot_table(index='category_floor_numeric', values='last_price', aggfunc='mean')
 .reset_index()
 .plot(title='Влияние типа этажа на цену', kind='scatter', x='category_floor_numeric', y='last_price', grid=True)
) 
```

### Зависисмость средней цены квадратного метра в г.Санкт-Петербург от расстояни до центра города

Выведем на экран информацию о средней стоимости квадратного метра от расстояния до центра города

```python
df['price_for_km'] = df['last_price'] / df['cityCenters_km']
sp_name = df[df['locality_name'] == 'санкт-петербург']
pivot = (
    sp_name
    .pivot_table(index='cityCenters_km', values=['price_for_km'], aggfunc=['mean'])
)

pivot.head(20)
```

Визуализируем графически изменение стоимости квадратного метра в пределах расстония от 200 метров до 3.5 км

```python
(
  sp_name
  .pivot_table(index='cityCenters_km', values=['price_for_km'], aggfunc=['mean'])
  .plot(title='Изменение средней цены от расстояния до центра в г.Санкт_Петербург (расстояние от 0.3 до 3.5 км)', grid=True, style='o-', xlim=(0.2, 3.5), figsize=(10, 5))
)
```

Видно что на данном отрезке максиммальная средняя цена, почти 80 млн за кв м, у квартир которые рассположены от центра города на дальше 200 метров. На отрезке от 400 до 450 метров цена растет и колеблется  колеблятся между 120 млн и 96 млн руб за квадратный метр. После этого сильное снижение до 24 млн за кв метр, и этот уровень сохраняется примерно до расстояния 1 км до центра. 
Высокая стоимость квадратного метра объясняется расположением в историческом центре и как правило это исторические здания с нестандартной планировкой: высокие потолки, большая площадь, большое количество комнат и т.д. Всплески в сторону роста можно объяснить тем что кроме расстояния от центра города, на стоимость влияют другие факторы.
