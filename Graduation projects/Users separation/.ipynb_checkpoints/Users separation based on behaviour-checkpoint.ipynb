### Выделяем сессии пользователей

Чтобы понимать и в последствии провести исследование (например по количеству событий в сессии), выделим пользовательские сессии на основе тайм-аута. Условия для выделения будут следующие:

- события совершает уникальный пользователь
- время между событиями не больше 30 минут

Если делать временной промежуток слишком маленьким, то вероятны ошибочные разделения одной сессии. Пользователи могут просто не успеть просмотреть объявление, либо они могут открыть фотографию и на что-то отвлечься. 

(Подход предложенный тимлидом)

```python

# Рассчитаем разницу времени между событиями
# для каждого пользователя
report = events_updated.groupby('user_id')['event_time'].diff()

# Числовое описание времени между событиями
print(report.describe())

# График времени между событиями
report.astype('timedelta64[s]').plot(
    kind='hist',
    bins=100,
    range=(500, 3600),
    figsize=(8, 5),
    legend=False
)
plt.title('Время между событиями')
plt.xlabel('время, с')
plt.ylabel('сессии')
plt.show()
```

Время между событиями изменяется в очень широком диапазоне - от долей секунды до 26 дней, стандартное отклонение 1 день 1 час 32 минуты. Однако, 75% значений укладывается в 3 минуты. Такая разница объясняется тем, что один пользователь может активно использовать приложение несколько минут подряд или повторно зайти спустя неделю — это нормальное поведение.

На гистограмме отображены временные интервалы между событиями в диапазоне от 0 до 3600 секунд (1 час). Видно, интервалы более 1800 секунд (30 минут) встречаются редко. 30 минут выберем в качкстве значения таймаута — если следующее событие одного пользователя происходит спустя более чем 30 минут от предыдущего, будем считать такое событие началом новой сессии.

```python
events_sorted=events_updated.sort_values(['user_id', 'event_time'])
events_sorted.head()
```

```python

g=(events_sorted.groupby('user_id')['event_time'].diff() > pd.Timedelta('30Min')).cumsum()

events_updated['session'] = events_updated.groupby(['user_id', g], sort=False).ngroup() + 1
print('Количество сессий - {}'.format(events_updated['session'].nunique()))
events_updated.head()
```

**Кол-во сессий 10368**

Посмотрим на пользователей с наибольшим количеством уникальных сессий

```python
events_updated.pivot_table(index='user_id', values='session', aggfunc='nunique').sort_values('session', ascending=False).head(10)
```

Исходя из лимита продолжительности сессии в 30 мин максимальное кол-во сессий на одного пользователя не превысило уровень 100

### Вычисление длительности сессии в минутах

```python
session_duration = events_updated.groupby('session')['event_time'].agg(['min','max']).reset_index()
session_duration['session_min'] = round((session_duration['max'] - session_duration['min']).dt.total_seconds()/60, 2)

display(session_duration.head())
print('Средняя продолжительность сессии - {}мин'.format(session_duration['session_min'].median()))
```

**Средняя продолжительность сессии около 6 минут.**

Посмотрим количество сессий состоящих из 1 события. В таких сессиях время не считается, т.е. равно 0 мин 0 сек.


```python
len(session_duration.loc[session_duration['session_min']==0])
```

```python
dataset_session = (
        events_updated.sort_values(by=['user_id', 'event_time'])
        .groupby(['user_id', 'session'])
        .agg(
            {
                'event_time': 'first',
                'source': 'first'
            }
        )
        .rename(columns={'event_time': 'session_start'})
        .reset_index()
    )
dataset_session.head()
```

```python
dataset_session = dataset_session.merge(
        session_duration[['session', 'session_min']],
        on=['session'],
        how='left',
    )
display(dataset_session.head())
dataset_session.info()
```

### Kоличество событий для пользователя

Подсчет количества событий и количества сессий в разрезе пользователей с указанием источников.

```python
event_number = (events_updated
                .groupby(['user_id'])
                .agg(
                    {
                        'event_name': 'count',
                        'session': 'nunique',
                        'source': 'first'
                    }
                )
                .rename(columns=
                    {
                        'event_name': 'event_number',
                        'session': 'session_number'
                    }
                )
                .reset_index()
)
event_number.sort_values('event_number', ascending=False).head()
```

Построим диаграммы размаха для кол-ва событий

```python
for x in ['event_number']:
    plt.figure(figsize=(18,9))
    ax1 = sns.boxplot(
        data=event_number, x=x, y='source',
        palette='YlGnBu',
        orient='h'
    )
    plt.xlim(0, 40)
    plt.title('Диаграмма размаха количества событий пользователя по источникам')
    plt.show()
```

Построим диаграммы размаха для кол-ва сессий

```python
for x in ['session_number']:
    plt.figure(figsize=(18,9))
    ax1 = sns.boxplot(
        data=event_number, x=x, y='source',
        palette='YlGnBu',
        orient='h'
    )
    plt.xlim(0, 7)
    plt.title('Диаграмма размаха количества сессий пользователя по источникам')
    plt.show()
    
```

- на диаграммах присутствуют выбросы, значит среднее количество нужно определять по медиане
- среднее (медианное) количество событий для пользователя и среднее количество сессий для разных исчтоников совпадают
- для пользователей из Yandex и Google больше характерны выбросы

Расчет среднего количества событий в сессии

```python
avoevent_number_per_session = (events_updated
                .groupby(['user_id', 'session'])
                .agg(
                    {
                        'event_name': 'count'
                    }
                )
                .rename(columns=
                    {
                        'event_name': 'event_number'
                    }
                )
                .reset_index()
)
avoevent_number_per_session.head()
```

Количество событий для пользователей и количество уникальных сессий

```python
plt.figure(figsize=(14, 9))
ax1 = sns.boxplot(
    data=avoevent_number_per_session['event_number'],
    palette='YlGnBu',
    orient='h'
)
plt.xlim(0, 25)
plt.title('Диаграмма размаха количества событий в сессии для пользователя')
plt.show()

avoevent_number_per_session['event_number'].describe()
```

### Retention Rate

добавление групп пользователей в датасет с событиями

```python
events_updated = events_updated.merge(
    users_session_time[['user_id', 'user_category']],
    on=['user_id'],
    how='left'
)
events_updated.head()
```

Создадим функцию, которая на входе будет принимать датасет с событиями пользователей evens_updated. Функция имеет возможность учитывать желаемый момент анализа, горизонт анализа. Также есть возможность разбиения датасета по интересующим признакам(в нашем случае по группам пользователей).

```python
def get_retention(
    profiles, # датасет с событиями
    observation_date, # момент анализа
    horizon_days, # горизонт анализа
    dimensions=[], # признак, задающий параметры разбиения датасета
    ignore_horizon=False, # возможность игнорировать горизонт анализа, по умолчанию отключено
):
    # делаем столбец user_category предзаданным в dimensions
    dimensions = ['user_category'] + dimensions
    
    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('event_time <= @last_suitable_acquisition_date')
    
    # вычисление времени первого события для уникальных пользователей
    first_ts = (profiles.sort_values(by=['user_id', 'event_time'])
        .groupby(['user_id'])
        .agg(
            {
                'event_time': 'first',
            }
        )
        .rename(columns={'event_time':'first_ts'})
        .reset_index()
    )

    # присоединение полученного значения времени первой сессии для пользователя к events_updated 
    result_raw = profiles.merge(
        first_ts,
        on=['user_id'],
        how='left',
    )
    
    # перевод типа данных столбцов со временем в недели
    result_raw['event_time'] = result_raw['event_time'].astype('datetime64[W]')
    result_raw['first_ts'] = result_raw['first_ts'].astype('datetime64[W]')
    
    # вычисляем лайфтайм каждой сессии в неделях
    result_raw['lifetime'] = (
        result_raw['event_time'] - result_raw['first_ts']
    ).dt.days / 7

    # строим таблицу удержания
    result_grouped = result_raw.pivot_table(
        index=dimensions, 
        columns='lifetime',
        values='user_id',
        aggfunc='nunique',
    )
    
    # вычисляем размеры когорт
    cohort_sizes = (
        result_raw.groupby(dimensions)
        .agg({'user_id': 'nunique'})
        .rename(columns={'user_id': 'cohort_size'})
    )
    
    # объединяем размеры когорт и таблицу удержания
    result_grouped = cohort_sizes.merge(
        result_grouped, on=dimensions, how='left'
    ).fillna(0)
    
    # делим данные таблицы удержания на размеры когорт
    result_grouped = result_grouped.div(result_grouped['cohort_size'], axis=0)

    # исключаем все лайфтаймы, превышающие горизонт анализа
    result_grouped = result_grouped[
        ['cohort_size'] + list(range(horizon_days))
    ]

    # восстанавливаем столбец с размерами когорт
    result_grouped['cohort_size'] = cohort_sizes

    # возвращаем таблицу удержания и сырые данные
    return result_raw, result_grouped
```

Вызываем функцию с разбивкой по группам и датам

```python
retention_raw, retention = get_retention(
    events_updated, datetime(2019, 10, 3).date(), 5, dimensions=['first_ts']
)
```

### Частота действий

Общее кол-во событий в датасете

```python
len(events_updated['event_name'])
```

```python
events_updated['event_name'].value_counts()
```

Подсчет количества различных событий для группы пользователей

```python
events_freq = (
    events_updated.groupby(['user_category', 'event_name'])
        .agg(
            {'event_name': 'count',
            }
        )
        .rename(columns={'event_name':'event_count'})
        .reset_index()
)
events_freq.head()
```

подсчет общего количества событий для группы пользователей

```python
freq_events_sum = (
    events_freq.groupby('user_category')
    .agg(
            {'event_count': 'sum'
            }
        )
    .rename(columns={'event_count':'event_count_sum'})
)

freq_events_sum.head()
```

**С ростом номера группы увеличивается кол-во событий**

```python
events_freq = events_freq.merge(
    freq_events_sum,
    on=['user_category'],
    how='left',
)
print('размер датасета - {}'.format(events_freq.shape))
events_freq.head()
```

вычисление частоты каждого события для группы пользователей

```python
events_freq['events_frequancy'] = round(100*events_freq['event_count'] / events_freq['event_count_sum'], 2)
events_freq.drop(['event_count', 'event_count_sum'], axis=1)
print('размер датасета - {}'.format(events_freq.shape))
events_freq.head()
```

## Проверка статистических гипотез

### Проверка гипотезы из технического задания

- Некоторые пользователи установили приложение по ссылке из yandex, другие — из google.
  - Гипотеза Н0: пользователи этих двух групп демонстрируют одинаковую конверсию в просмотры контактов
  - Гипотеза Н1: пользователи этих двух групп демонстрируют разную конверсию в просмотры контактов

```python
users_yandex=events_updated.query('source=="yandex"')['user_id'].unique().tolist()
print('Кол_во пользователей установивших приложение из Yandex, равно:', len(users_yandex))
print('Кол-во пользователей установивших приложение из Yandex совершивших целевое действие:', events_updated.query('user_id==@users_yandex and event_name=="contacts_show"')['user_id'].nunique())
convertion_yandex_=events_updated.query('user_id==@users_yandex and event_name=="contacts_show"')['user_id'].nunique()/len(users_yandex)    
print(f'Конверсия в ЦС пользователей Yandex {round(convertion_yandex_*100, 2)}%')    
    
    
users_google=events_updated.query('source=="google"')['user_id'].unique().tolist()
print('Кол_во пользователей установивших приложение из Google, равно:', len(users_google))
print('Кол-во пользователей установивших приложение из Google совершивших целевое действие:', events_updated.query('user_id==@users_google and event_name=="contacts_show"')['user_id'].nunique())
convertion_google_=events_updated.query('user_id==@users_google and event_name=="contacts_show"')['user_id'].nunique()/len(users_google)    
print(f'Конверсия в ЦС пользователей Google {round(convertion_google_*100, 2)}%')    

```

**Проверка гипотезы из технического задания**

- Данные для теста выглядят следующим образом:

  - Кол-во Yandex пользователей = 1934

  - Кол-во ЦС у Yandex пользователей = 478

  - Кол-во Google пользователей = 1129

  - Кол-во ЦС у Google пользователей = 275

```python
def stat_test(successes, trials, alpha):  

    # alpha = .05 # критический уровень статистической значимости

    # successes = np.array([478, 275])
    # trials = np.array([1934, 1129])

    alpha = alpha
    successes = successes
    trials = trials

    # пропорция успехов в первой группе:
    p1 = successes[0]/trials[0]

    # пропорция успехов во второй группе:
    p2 = successes[1]/trials[1]

    # пропорция успехов в комбинированном датасете:
    p_combined = (successes[0] + successes[1]) / (trials[0] + trials[1])

    # разница пропорций в датасетах
    difference = p1 - p2


    # считаем статистику в ст.отклонениях стандартного нормального распределения
    z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/trials[0] + 1/trials[1]))

    # задаем стандартное нормальное распределение (среднее 0, ст.отклонение 1)
    distr = st.norm(0, 1) 

    p_value = (1 - distr.cdf(abs(z_value))) * 2

    print('p-значение: ', p_value)

    if p_value < alpha:
        print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
    else:
        print(
            'Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными'
        )

    
alpha = .05
successes = [events_updated.query('user_id==@users_yandex and event_name=="contacts_show"')['user_id'].nunique(), events_updated.query('user_id==@users_google and event_name=="contacts_show"')['user_id'].nunique()]
trials = [len(users_yandex), len(users_google)]

stat_test(successes, trials, alpha)  
```

**Вывод:** 

В результате проверки статистической гипотезы о равенстве долей z-критерием оказалось, что разница в конверсии между пользователями, пришедшими из поисковика Yandex, и теми, кто пришёл из Google, не является статистически значимой.
